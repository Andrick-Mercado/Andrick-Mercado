```python
import time
import math
class studentSolution(Sudoku):
    """ This subclass is almost all yours
    The only restriction is that you need to have a solution() method which will return the solution path
    I recommend that you treat it mostly like a dummy function and just call your function, but it is up to you
    """
    def __init__(self,board=False,test=False):
        # Constructor for subclass
        # Add to it if you need to 
        super().__init__(board,test)
    
    ## Your code ###
    # Add methods as needed to support the solution
    
    
    def solution(self):
        
        # This is mostly for testing. You can remove it if it doesn't work for you
        solved_board = self.board.copy()
        """ 
        Once completed this should return your solved board
        """
        
        ### Your code here
        #I prfer using arrays instead of a dataframe hence why I do this work here
        #below i just cleaned the data to make it work every time no matter the inputs
        board_dict = pd.DataFrame.from_dict(solved_board)#MAKE it a panda dataframe, in case of array or regular dict
        board_dict[np.isnan( board_dict )] = 0 #replace np.nan for zeros
        board_dict = board_dict.values.tolist()#make dictionary to list, only the values
        board_dict = [[int(y) for y in x] for x in board_dict]#make the numbers integers
        solved_board = board_dict

        keys = ['A','B','C','D','E','F','G','H','I']
        dataFrame = dict(zip(keys, [list(i) for i in zip(*board_dict)]))#make the list back to a dictionary
        board_dict = pd.DataFrame.from_dict(dataFrame)
        
        self.solve_board(solved_board)#solve the board once we have cleaned the data
        
        dataFrame = dict(zip(keys, [list(i) for i in zip(*solved_board)]))#make the solved list to dictionary
        solved_board = pd.DataFrame.from_dict(dataFrame)
        
        #### End your code
        
        return solved_board
    
    def solve_board(self, board):
        find = self.find_empty_cell( board)
        if not find:#recursive base case: when board is solved
            return True
        
        row, col = find#if board is not found we find an empty cell or zero and get its position

        for i in range(1,10):#try all possible values in suduko 1 to 9
            if self.is_valid_number( board, i, (row, col)):
                board[row][col] = i

                if self.solve_board(board):#recursion backtracking
                    return True

                board[row][col] = 0#if backtracking fails we set value back to 0 or in other words reset

        return False#default case since we dont have a solved board yet

    ''' 
        We take in the board a number which is the number we attempt to put at a given position
        here is were we implement the constraint algorithm (validation)
        returns true if the given number is valid on the position given
        returns false otherwise
    '''
    def is_valid_number(self,board, number_attempt, position):
        # check if the valuie is already in the row false if num is invalid
        for i in range(len(board[0])):
            if board[position[0]][i] == number_attempt and position[1] != i:
                return False

        # check if the valuie is already in the column false if num is invalid
        for i in range(len(board)):
            if board[i][position[1]] == number_attempt and position[0] != i:
                return False

        # get the location of the box
        #essentially we have 9 total boxes and based on the given position we calculate the 
        #position of the box we want to begin from position (0,0)of a 3 * 3 matrix, hence why
        #we cut the number and multiply by three
        box_position_x = math.floor( position[1] / 3)*3
        box_position_y = math.floor( position[0] / 3)*3

        for i in range(box_position_y, box_position_y + 3):#get the box indeces for rows: n to n+3
            for j in range(box_position_x , box_position_x + 3):#get the box indeces for cols: n to n+3
                if board[i][j] == number_attempt and (i,j) != position:#check if the new value violates the box rules
                    return False

        return True#if no violations found in rows, cols, and box


    def find_empty_cell(self, board):#return index of first occurance of a zero
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == 0:
                    return i, j 
        return None

    
''' TESTING ZONE'''
array_board1 = [ [7,8,0,4,0,0,1,2,0],
                [6,0,0,0,7,5,0,0,9],
                [0,0,0,6,0,1,0,7,8],
                [0,0,7,0,4,0,2,6,0],
                [0,0,1,0,5,0,9,3,0],
                [9,0,4,0,6,0,0,0,5],
                [0,7,0,3,0,0,0,1,2],
                [1,2,0,0,0,7,4,0,0],
                [0,4,9,2,0,6,0,0,7] ]

array_board2 = [ [0,0,5,3,0,0,0,0,0],
                [8,0,0,0,0,0,0,2,0],
                [0,7,0,0,1,0,5,0,0],
                [4,0,0,0,0,5,3,0,0],
                [0,1,0,0,7,0,0,0,6],
                [0,0,3,2,0,0,0,8,0],
                [0,6,0,5,0,0,0,0,9],
                [0,0,4,0,0,0,0,3,0],
                [0,0,0,0,0,9,7,0,0] ]

board_dict1= {'A':[3,np.nan,1,7,np.nan,np.nan,5,np.nan,np.nan],
             'B':[6,4,np.nan,np.nan,np.nan,8,np.nan,3,np.nan],
             'C':[5,np.nan,7,2,np.nan,3,np.nan,np.nan,np.nan],
             'D':[np.nan,np.nan,np.nan,np.nan,6,np.nan,8,np.nan,5],
             'E':[9,6,5,8,np.nan,np.nan,np.nan,7,4],
             'F':[2,1,8,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],
             'G':[7,np.nan,9,np.nan,np.nan,4,6,np.nan,1],
             'H':[np.nan,np.nan,6,np.nan,np.nan,7,2,np.nan,np.nan],
             'I':[8,np.nan,4,np.nan,2,np.nan,np.nan,np.nan,7]}

board_dict2 = {'A':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],
              'B':[1,9,np.nan,np.nan,8,np.nan,3,np.nan,np.nan],
              'C':[np.nan,6,np.nan,np.nan,5,4,np.nan,9,np.nan],
              'D':[np.nan,np.nan,2,np.nan,np.nan,np.nan,np.nan,np.nan,5],
              'E':[np.nan,np.nan,np.nan,np.nan,6,np.nan,np.nan,3,4],
              'F':[np.nan,7,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],
              'G':[5,np.nan,np.nan,8,np.nan,np.nan,np.nan,np.nan,np.nan],
              'H':[np.nan,np.nan,1,np.nan,np.nan,np.nan,9,np.nan,6],
              'I':[4,np.nan,np.nan,7,2,np.nan,np.nan,5,np.nan]}

print("Four test 2 array form others are dictionaries")
print("-----------------------------")
print("EASY ARRAY")
print("-----------------------------")

start = time.time()
test = studentSolution(array_board1)
sol_board = test.solution()
print(sol_board)
print("Time elapsed: ", (time.time() -start))

print("-----------------------------")
print("HARD ARRAY")
print("-----------------------------")

start = time.time()
test2 = studentSolution(array_board2)
sol_board = test2.solution()
print(sol_board) 
print("Time elapsed: ", (time.time() -start))

print("-----------------------------")
print("EASY DICTIONARY")
print("-----------------------------")

start = time.time()
test3 = studentSolution(board_dict1)
sol_board = test3.solution()
print(sol_board) 
print("Time elapsed: ", (time.time() -start))

print("-----------------------------")
print("HARD DICTIONARY")
print("-----------------------------")

start = time.time()
test4 = studentSolution(board_dict2)
sol_board = test4.solution()
print(sol_board) 
print("Time elapsed: ", (time.time() -start))

print("-----------------------------")
```
