#P01

Derive the Taylor Series for 
$$ f(x) = e^x $$ 

Hence by Taylor series,
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since 
$$ f(x) = f'(x) = f''(x)=.... = e^x$$ 
and, $$ f(0) = f'(0) = f''(0)=.... = e^0 = 1 $$
Hence, 
$$ f(x) = \sum_{n=0}^{\infty}\frac{1}{n!}x^n =\sum_{n=0}^{\infty}\frac{x^n}{n!} $$

#P06

Derive the Taylor Series for 
$$ h(x) = \cos(x) $$ 

Hence by Taylor series,
$$h(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$h(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since, 

\begin{equation} 
\begin{split}
 h(0) &=   cos(0)&=1 \\
 h'(0) &=  -sin(0)&=0  \\
 h''(0) &= -cos(0) &= -1\\
 h'''(0) &= sin(0) &= 0 \\
 h^{(4)}(0) &= cos(0) &= 1
\end{split}
\end{equation}

As wecontinue with further derivatives the cycle stays the same and we have the cycle $\{1,0,-1,0,...\}$ and so on.
Hence since it alternates we get,
$$ h(x) = 1 - \frac{x^2}{2!}+... = \sum_{n=0}^{\infty} (-1)^n\cdot \frac{x^{2n}}{(2n)!}$$

**Score:** 6/10
* Check your $\LaTeX$

#P07

```
n_terms = 7
x_vals = np.linspace(-2,2,100)

y_vals = np.zeros([n_terms,len(x_vals)])
plt.plot(x_vals, np.cos(x_vals), label="$h(x) = \cos(x)$")
for n in range(n_terms):
    if n == 0:
        y_vals[n,:] = np.ones(len(x_vals))
    else:
        new_term = (-1)**n * x_vals**(2*n) / np.math.factorial(2*n)
        y_vals[n,:] = y_vals[n-1,:] + new_term

plt.plot(x_vals,y_vals[2,:], label='$y=\sum_{n=0}^{2}$')
plt.plot(x_vals,y_vals[4,:], label='$y=\sum_{n=0}^{4}$')
plt.plot(x_vals,y_vals[6,:], label='$y=\sum_{n=0}^{6}$')

plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-1.1,1.1]) 
plt.grid(alpha = 0.618)
plt.legend(shadow = True, loc = 8)
plt.hlines(0, -2, 2,color = 'k')
plt.vlines(0, -2, 2,color = 'k')
```

#P08

Derive the Taylor Series for 
$$ j(x) = \sin(x) $$ 

Hence by Taylor series,
$$j(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$j(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since, 

\begin{equation} 
\begin{split}
 j(0)&=sin(0)&=0 \\
 j'(0)&=cos(0)&=1  \\
 j''(0) &= -sin(0) &= 0\\
 j'''(0) &= -cos(0) &= -1 \\
 j^{(4)}(0) &= sin(0) &= 0
\end{split}
\end{equation}
From this we get:
$$ j(x) = x-\frac{x^3}{3!}+\frac{x^5}{5!}-...= \sum_{n=0}^\infty\frac{(-1)^n}{(2n+1)!}\cdot x^{2n+1} $$
