The golden ratio is observed throughout nature. We can compute the golden ratio as

$$
\lim_{n\rightarrow \infty}\frac{F(n+1)}{F(n)}=\phi,
$$
where $\phi$ is the golden ratio. We can use our function above to asymptotically approximate $\phi\approx 1.61803$.

`````julia
ratios = []
x = 1:10
for i in x
    push!(ratios, Fibonacci(i+1)/Fibonacci(i))
    println(i,"\t",ratios[i])
end
`````

`````julia
plot(x,ratios, marker = 'o', linewidth=0)
`````

# Problem 0

An important function in physics is the [Airy function](https://en.wikipedia.org/wiki/Airy_function), which solves
$$\frac{d^2y}{dx^2}-xy=0,$$
given by the Taylor Series

$$
y_{1}(x) = 1 + \sum_{m=1}^{\infty}c_{m}x^{3m},\\
\text{where}\quad c_{m} = \prod_{l=1}^{m} \frac{1}{(3l-1)(3l)}.
$$

Note the coefficients $c_{m}$ follow the pattern

\begin{align}
c_{1} = & \prod_{l=1}^{1} \frac{1}{(3l-1)(3l)} = \frac{1}{2\cdot 3} \approx 0.16667\\
c_{2} = & \prod_{l=1}^{2} \frac{1}{(3l-1)(3l)} = \frac{1}{2\cdot 3 \cdot 5 \cdot 6} \approx 0.0055556
\end{align}
