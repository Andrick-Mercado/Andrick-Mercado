**Question 3**

```julia
function solution3(a)
    s = 0.0
    for x in 0:a
        s+=1/(3^x)
    end
    return s
end
println( "S(1000000) = ",solution3(1000000) )
```

```julia
#T = 999999:0.1:1000000
#plot(T, solution3, xlabel="n", ylabel="y", label="S(n)")
T = 0.0: 100
scatter(T, solution3, xlabel="n", ylabel="y", label="S(n)")
ylims!((1.49,1.501))
```

**Question 5**

```julia
function solution5(a)
    s = 0.0
    for x in 1:a
        s+=((-1)^x)/((2*x)-1)
    end
    return -4*s
end
println( "S(1000000) = ",solution5(1000000) )
```

```julia
#T = 999999:1:1000000
#plot(T, solution5, xlabel="n", ylabel="y", label="S(n)")
T = 0.0: 100
scatter(T, solution5, xlabel="n", ylabel="y", label="S(n)")
ylims!((2,4))
```
