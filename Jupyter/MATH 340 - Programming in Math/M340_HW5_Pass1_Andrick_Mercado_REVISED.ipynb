#P01

Derive the Taylor Series for 
$$ f(x) = e^x $$ 

Hence by Taylor series,
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since 
$$ f(x) = f'(x) = f''(x)=.... = e^x$$ 
and, $$ f(0) = f'(0) = f''(0)=.... = e^0 = 1 $$
Hence, 
$$ f(x) = \sum_{n=0}^{\infty}\frac{1}{n!}x^n =\sum_{n=0}^{\infty}\frac{x^n}{n!} $$
 

#P02

```
def S(n):
  return sum([ ((n**i)/(np.math.factorial(i))) for i in range(n+1)])
def f(x):
  return np.exp(x)

fig, axs = plt.subplots(2, 2)

n = [*range(2)]
y = [S(_) for _ in n]
axs[0, 0].plot(n, y)
axs[0,0].grid(alpha = 0.618)
axs[0,0].set_title('$y=\sum_{n=0}^{1}\fract{x^n}{n!}$')

n = [*range(3)]
y = [S(_) for _ in n]
axs[0, 1].plot(n, y, 'tab:orange')
axs[0,1].grid(alpha = 0.618)
axs[0, 1].set_title('$y=\sum_{n=0}^{2}\fract{x^n}{n!}$')

n = [*range(4)]
y = [S(_) for _ in n]
axs[1, 0].plot(n, y, 'tab:green')
axs[1,0].grid(alpha = 0.618)
axs[1, 0].set_title('$y=\sum_{n=0}^{3}\fract{x^n}{n!}$')

n = [*range(-2,3)]
y = [f(_) for _ in n]
axs[1, 1].plot(n, y, 'tab:red')
axs[1,1].grid(alpha = 0.618)
axs[1, 1].set_title('y=$e^x$')
fig.tight_layout()
```

#P04

Derive the Taylor Series for 
$$ f(x) = e^{x^2} $$ 

Hence by Taylor series,
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since 

$$e^x = \sum_{n=0}^{\infty}\frac{x^n}{n!} $$
we just say,
$$ f(x) = e^{x^2}=\sum_{n=0}^{\infty}\frac{x^{2n}}{n!}  $$

#P05

```
n_terms = 4
x_vals = np.linspace(-2,2,100)

y_vals = np.zeros([n_terms,len(x_vals)])
plt.plot(x_vals, np.exp(x_vals**2), label="$g(x) = e^{x^2}$")

n = 1
new_term = ((x_vals**(n*2))/(np.math.factorial(n)))
y_vals[n,:] = y_vals[n-1,:] + new_term
plt.plot(x_vals,y_vals[n,:], label='$y=\sum_{n=0}^{1}$')

n = 2
new_term = ((x_vals**(n*2))/(np.math.factorial(n)))
y_vals[n,:] = y_vals[n-1,:] + new_term
plt.plot(x_vals,y_vals[n,:], label='$y=\sum_{n=0}^{2}$')

n = 3
new_term = ((x_vals**(n*2))/(np.math.factorial(n)))
y_vals[n,:] = y_vals[n-1,:] + new_term
plt.plot(x_vals,y_vals[n,:], label='$y=\sum_{n=0}^{3}$')

plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha = 0.618)
plt.legend(shadow = True, loc = 0)
plt.hlines(0, -2, 2,color = 'k')
plt.vlines(0, 1, 10,color = 'k')
```

#P06

Derive the Taylor Series for 
$$ h(x) = \cos(x) $$ 

Hence by Taylor series,
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since, 

\begin{equation} 
\begin{split}
 f(0) &=   cos(0)&=1 \\
 f'(0) &=  -sin(0)&=0  \\
 f''(0) &= -cos(0) &= -1\\
 f'''(0) &= sin(0) &= 0 \\
 f^{(4)}(0) &= cos(0) &= 1
\end{split}
\end{equation}

As wecontinue with further derivatives the cycle stays the same and we have the cycle $\{1,0,-1,0,...\}$ and so on.
Hence since it alternates we get,
$$ f(x) = 1 - \frac{x^2}{2!}+... = \sum_{n=0}^{\infty} (-1)^n\cdot \frac{x^{2n}}{(2n)!}$$

#P07

```
n_terms = 7
x_vals = np.linspace(-2,2,100)

y_vals = np.zeros([n_terms,len(x_vals)])
plt.plot(x_vals, np.cos(x_vals), label="$h(x) = \cos(x)$")
for n in range(n_terms):
    if n == 0:
        y_vals[n,:] = np.ones(len(x_vals))
    else:
        new_term = (-1)**n * x_vals**(2*n) / np.math.factorial(2*n)
        y_vals[n,:] = y_vals[n-1,:] + new_term

plt.plot(x_vals,y_vals[2,:], label='$y=\sum_{n=0}^{2}$')
plt.plot(x_vals,y_vals[4,:], label='$y=\sum_{n=0}^{4}$')
plt.plot(x_vals,y_vals[6,:], label='$y=\sum_{n=0}^{6}$')

plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-1.1,1.1]) 
plt.grid(alpha = 0.618)
plt.legend(shadow = True, loc = 8)
plt.hlines(0, -2, 2,color = 'k')
plt.vlines(0, -2, 2,color = 'k')
```

#P08

Derive the Taylor Series for 
$$ j(x) = \sin(x) $$ 

Hence by Taylor series,
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(a)}{n!}(x-a)^n$$
Next $x_0 = 0 = a$
$$f(x)= \sum_{n=0}^{\infty}\frac{f^{n}(0)}{n!}x^n$$
Now since, 

\begin{equation} 
\begin{split}
 f(0)&=sin(0)&=0 \\
 f'(0)&=cos(0)&=1  \\
 f''(0) &= -sin(0) &= 0\\
 f'''(0) &= -cos(0) &= -1 \\
 f^{(4)}(0) &= sin(0) &= 0
\end{split}
\end{equation}
From this we get:
$$ j(x) = x-\frac{x^3}{3!}+\frac{x^5}{5!}-...= \sum_{n=0}^\infty\frac{(-1)^n}{(2n+1)!}\cdot x^{2n+1} $$

#P09

```
sum_of_nums = 7
range_of_x = np.linspace(-2,2,100)

y_vals = np.zeros([sum_of_nums,len(range_of_x)])#array of values
plt.plot(range_of_x, np.sin(range_of_x), label="$j(x) = \sin(x)$")
for n in range(sum_of_nums):
    if n == 0:
        y_vals[n,:] = range_of_x
    else:
        new_term = (-1)**n * range_of_x**((2*n)+1) / np.math.factorial((2*n)+1)
        y_vals[n,:] = y_vals[n-1,:] + new_term

plt.plot(range_of_x,y_vals[1,:], label='$y=\sum_{n=0}^{1}$')

plt.plot(range_of_x,y_vals[3,:], label='$y=\sum_{n=0}^{3}$')

plt.plot(range_of_x,y_vals[5,:], label='$y=\sum_{n=0}^{5}$')

plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-1.1,1.1]) 
plt.grid(alpha = 0.618)
plt.legend(shadow = True, loc = 0)
plt.hlines(0, -2, 2,color = 'k')
plt.vlines(0, -2, 2,color = 'k')
```
