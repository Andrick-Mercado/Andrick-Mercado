Using the Taylor series, prove the identities.  For example, given that

$$
e^{x} = \sum_{j=0}^{\infty} \frac{x^{j}}{j!}
$$

we can show that 

$$
\frac{d}{dx}e^{x} = e^{x}
$$

by seeing that 

\begin{align}
\frac{d}{dx}e^{x} = & \sum_{j=0}^{\infty} \frac{d}{dx} \frac{x^{j}}{j!}\\
= & \sum_{j=1}^{\infty} j\frac{x^{j-1}}{j!}\\
= & \sum_{j=1}^{\infty} \frac{x^{j-1}}{(j-1)!}\\
= & \sum_{k=0}^{\infty} \frac{x^{k}}{k!}, ~ k = j-1\\
= & e^{x}.
\end{align}

    show that
    $$
    \frac{d}{dx}\cos(x) = -\sin(x).
    $$
    

$\LaTeX$ **Solution goes here**

Let us begin by stating the following,
$$\frac{d}{dx}\cos(x) = -\sin(x)$$
Now the proof, 
\begin{equation} 
\begin{split}
 \frac{d}{dx} \cos(x) &=\frac{d}{dx}\left(\sum_{j=0}^{\infty}\frac{(-1)^{j}x^{2j}}{(2j)!}\right) \\
 -\sin(x) &= \sum_{j=1}^{\infty}\frac{(-1)^{j}\cdot2j\cdot x^{2j-1}}{(2j)!}\\
 &=  \sum_{j=1}^{\infty}\frac{(-1)^{j}\cdot2j\cdot x^{2j-1}}{(2j)(2j-1)!}\\
 &=  \sum_{j=1}^{\infty}\frac{(-1)^{j}\cdot x^{2j-1}}{(2j-1)!}
\end{split}
\end{equation}

Now we shift $j$ to $0$,
$$ \sum_{j=0}^{\infty}\frac{(-1)^{j+1}\cdot x^{2j+1}}{(2j+1)!} $$
We further simplify the series,
$$ \sum_{j=0}^{\infty}\frac{(-1)^{j+1}\cdot x^{2j+1}}{(2j+1)!} = \sum_{j=0}^{\infty}\frac{(-1)^{j}\cdot(-1)^1\cdot x^{2j+1}}{(2j+1)!} $$
Since $(-1)^1$ is a constant we can move it,
$$ (-1)\cdot\sum_{j=0}^{\infty}\frac{(-1)^{j}\cdot x^{2j+1}}{(2j+1)!} $$
Given the fact that,
$$\sin(x) = \sum_{j=0}^{\infty}\frac{(-1)^{j}x^{2j+1}}{(2j+1)!} $$
We say that,
$$ -sin(x) = (-1)\cdot\sum_{j=0}^{\infty}\frac{(-1)^{j}x^{2j+1}}{(2j+1)!}$$
Hence, 
$$\frac{d}{dx}\cos(x) = -\sin(x).$$

#P4 (10 points)

* Using, $$ \frac{1}{1-t} = \sum_{j=0}^{\infty}t^{j}, ~ ~ |t|<1 \\ \text{and}\qquad\ln(1+x) = \int_{0}^{x}\frac{dt}{1+t}, ~ ~ x \in (-1,\infty) $$
    show  that $$  \ln(1+x) = \sum_{j=1}^{\infty}(-1)^{j-1}\frac{x^{j}}{j}. $$
    Remember, when we have convergence, we have  $$\int_{0}^{x} \sum_{j=0}^{\infty} f_{j}(t) dt = \sum_{j=0}^{\infty} \int_{0}^{x} f_{j}(t)dt. $$

$\LaTeX$ **Solution goes here**

First we state the following equivalence,
$$ \ln(1+x)= \int_{0}^{x}\frac{dt}{1+t}= \int_{0}^{x}\frac{dt}{1-(-t)} $$
Now we expand the integral,
$$ \int1+(-t)+(-t)^2+(-t)^3++(-t)^4...dt $$
which is equivalent to,
$$ \int\sum_{n=0}^\infty(-t)^n \qquad,|-t|<1$$ 
Using the following equivalence we expand,
$$ \int1-t+t^2-t^3+t^4...dt  $$
which is equivalent to,
$$ \int\sum_{n=0}^\infty t^{n+1}\frac{(-1)^{n}}{n+1}dt \qquad,|t|<1 $$
Since 1 is just a constant we let $t =0$,
\begin{equation} 
\begin{split}
 \ln(1+0) &=  b + 0 - \frac{1}{2}0^2+...-0  \\
 \ln(1)&=b \\
 0&= b
\end{split}
\end{equation}
Now we can say,
$$ \ln(1+x) = t-\frac{1}{2}t^2+\frac{1}{3}t^3-\frac{1}{4}t^4+.... $$
From this we conclude it is an alternating series hence,
$$ \ln(1+x) =\sum_{n=0}^\infty t^{n+1}\frac{(-1)^{n}}{n+1} = \sum_{n=0}^{\infty}(-1)^{n}\frac{t^{n+1}}{n+1} $$

Now we say $n = j$ and $t = x$,
$$ \ln(1+x) = \sum_{j=0}^{\infty}(-1)^{j}\frac{x^{j+1}}{j+1} $$
Now lets looked at the shifted series,
$$ \sum_{j=1}^{\infty}(-1)^{j-1}\frac{x^{j}}{j} = x-\frac{1}{2}x^2+\frac{1}{3}x^3-\frac{1}{4}x^4+....  = \sum_{j=0}^{\infty}(-1)^{j}\frac{x^{j+1}}{j+1} $$
Hence,
$$ \ln(1+x) = \sum_{j=1}^{\infty}(-1)^{j-1}\frac{x^{j}}{j}. $$

`````python
# Solution to Problem 5:  
def S(n):
  return sum([(-1)**(i+1)/i for i in range(1,n+1)])    

print('Estimate the value of ln(2) at 10^1 =',"{:.6f}".format(S(10)) )
print('Estimate the value of ln(2) at 10^2 =',"{:.6f}".format(S(100)) )
print('Estimate the value of ln(2) at 10^3 =', "{:.6f}".format(S(1000)) )
print('Estimate the value of ln(2) at 10^4 =', "{:.6f}".format(S(10000)) )

''' TABLE '''
fig, ax = plt.subplots()
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
df = pd.DataFrame([['$10^1 $',0.645635],
                   ['$10^2$', 0.688172], 
                   ['$10^3$',0.692647],
                   ['$10^4$', 0.693097]],
                   columns=['n', 'S(n)'])
ax.table(cellText=df.values, colLabels=df.columns, loc='center')
fig.tight_layout()
plt.subplots_adjust(left=0.3, top=1)
plt.show()

''' plot '''
plt.figure(figsize=(10,5))
plt.title("Approximation of ln(2)")
plt.plot(0,0.645635 , 'o',mfc = 'none', label = '$y=\sum_{n=0}^{10^1}$')
plt.plot(1, 0.688172 , 'o',mfc = 'none', label = '$y=\sum_{n=0}^{10^2}$')
plt.plot(2,0.692647, 'o',mfc = 'none', label = '$y=\sum_{n=0}^{10^3}$')
plt.plot(3, 0.693097 , 'o',mfc = 'none', label = '$y=\sum_{n=0}^{10^4}$')

plt.hlines(np.log(2),0,1E4, 'k', linestyle='dashed', label=r'$y=\ln(2)$')
plt.xlabel('$i$')
plt.ylabel('$\mathcal{E}_i$')
plt.grid(alpha = 0.618)
plt.legend(shadow = True, loc = 4)
plt.ylim([0.64,0.7]) 
plt.xlim([-.1,3.1]) 
`````

An important function in physics is the [Airy function](https://en.wikipedia.org/wiki/Airy_function), which solves
$$\frac{d^2y}{dx^2}-xy=0,$$
given by the Taylor Series

$$
y_{1}(x) = 1 + \sum_{m=1}^{\infty}c_{m}x^{3m},\\
\text{where}\quad c_{m} = \prod_{l=1}^{m} \frac{1}{(3l-1)(3l)}.

Note the coefficients $c_{m}$ follow the pattern

\begin{align}
c_{1} = & \prod_{l=1}^{1} \frac{1}{(3l-1)(3l)} = \frac{1}{2\cdot 3}\\

`````python
# func c(m)
def c(m):# product will become 1 if m is not greater than zero
  count = 1
  product = 1
  while (count<=m):
    product = product * (1/((3*count-1)*(3*count)))
    count+=1
  return product
print("Testing c(1) = ",c(1), "and c(2) = ",c(2))
`````

`````python
# func y1(n,x)
def y1(n,x):
  return 1+sum([c(m)*x**(3*m) for m in range(1,n+1)])
`````
